# Exploring the Relationship Between Life Expectancy and GDP

With this data set, we will look at six different countries and their GDPs over a 15 year period. The countries in quesiton are Chile, Zimbabwe, Germany, China, the United States of America and Mexico. Our goal is to look for any relationships we can find between life expectancy and GDP and then try to understand why that relationship may exist.

- Conclusion: The Relationship between Life Expectancy and GDP

## Exploring the Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

Let's take a look at our data.

```python
df = pd.read_csv('all_data.csv')
df.head(96)
#df.describe(include = 'all')
```

```python
df.Country.value_counts()
```

We can see that we have an equal number of entries for all countries in our data.  Lets see how the GDP has changed over the years for each country.

```python
g = sns.FacetGrid(df, col="Country", col_wrap = 3, palette = 'colorblind')
g.map(sns.scatterplot, "Year", "GDP", alpha=.9)
g.add_legend()
plt.show()
```

```python
sns.scatterplot(x = 'Year', y = "GDP", data = df, hue = "Country")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```

Here we can see that most countries show some level of upward trend, though it is difficult to tell for some, specifically Zimbabwe. Some countries also show a downward trend after 2008.

### Effect of Economic Crisis
    
Visable in the above graph is a small dip in GDP following the year 2008, mainly in the Americas and Germany, with smaller effects on the other countries.  This is a result of the global economic crisis that charactorized 2008 that stemmed from irresponsible banking and investing practices of United States financial giants. This crisis, called the Great Recession, spread throughout Europe and had a heavy impact on the automotive industry, which hit both European, American and Asian industries.

The Great Recession had a much smaller impact on the GDPs of China and Zimbabwe, both of which did not have extensive banking or automotive industries at the time of the recession.

Let's see how life expectancy has changed over the years

```python
h = sns.FacetGrid(df, col="Country", col_wrap = 3,)
h.map(sns.scatterplot, "Year", "Life expectancy at birth (years)", alpha=.9)
plt.show()
```

```python
sns.scatterplot(x = 'Year', y = "Life expectancy at birth (years)", data = df, hue = "Country")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```

We can see that Zimbabwe took a major downturn in life expectancy in the early 2000s and is making their way back up to a higher life expectancy.  Otherwise all the other countries seem to have a generally upward trend.  Chile and Mexicao seem to have a noticable downward bump between 2005 and 2010.

### Comment on Zimbabwe
Noticeably, Zimbabwe has suffered a significant loss of life expectancy in the early 2000's and has begun to recover in recent years.  The data does not show life expectancy before 2000 or after 2015, so it is not possible with the given information to determine if this trend was visable prior to 2000.  However, with some research, it is found that Zimbabwe has had considerable political instability in the early 2000s which has lead to violence throughout the country.  Zimbabwe also had large spread housing insecurity and HIV/AIDS outbreaks. All of these factors contribute to the low life expectancy relatively to the other countries being considered.

In an effort to gain further insights, I will now work with the data excluding Zimbabwe. I am hopeful this will make it easier to compare the GDP and Life Expectancies of the other countries.

```python
new_df = df.drop(range(80,96))
new_df.Country.value_counts()
```

```python
sns.scatterplot(x = 'Year', y = "GDP", data = new_df, hue = "Country")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```

```python
sns.scatterplot(x = 'Year', y = "Life expectancy at birth (years)", data = new_df, hue = "Country")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```

```python
sns.scatterplot(x = 'GDP', y = "Life expectancy at birth (years)", data = new_df, hue = "Country")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```

Now it is easier to see the data for the five other countries, excluding Zimbabwe. We can see the same trends as we found before, but can see them more clearly now.

## Conclusion: Relationship between Life Expectancy and GDP

From the preceeding graphs, it is clear that there just because a country has the highest GDP does not mean it has the highest life expectancy. While the United States has had the advantage in GDP for 15 years and China's GDP has surged ahead in recent years, neither China or the United States leads in life expectancy. Germany and Chile have relatively lower GDPs, yet pull ahead of the larger GDP countries when it comes to life expectancy.  This means that GDP is not nessecarily an indicator of long living citizenship, so there must be other factors in play that are not displayed in the given data.  Some of these factors could be social welfare systems, low crime rates, access to higher education and more.
